```
def xsqr(x):
    return x**2

# Exact
#x^3/3 + c
```

```
x = np.linspace(3, 10, 100)
y = xsqr(x)
```

```
plt.plot(x,y)
plt.fill_between(x, y, 0, color='C0')
```

```
a = 3
b = 10

def exact_soln(x):
    return x**3/3 

exact = exact_soln(b) - exact_soln(a)
```

```
#https://en.wikipedia.org/wiki/Trapezoidal_rule
def trapezoidal(a,b,func):
    return (b-a)*(func(a) + func(b))/2
```

```
trapezoidal(a, b, xsqr), exact
```

```
def simpson(a, b, func):
    h = (b-a)/2
    return h/3*(func(a) + 4*func((a+b)/2) + func(b))

def simpson_3by8(a, b, func):
    return ((b-a)/8)*(func(a) + 3*func((2*a+b)/3) + 3*func((a+2*b)/3) + func(b))
```

```
simpson(a,b,xsqr), simpson_3by8(a,b,xsqr), exact
```

```
x = np.linspace(3,10,10000)
y = xsqr(x)

plt.plot(x,y)
```

```
width = np.diff(x)
height = 0.5*(y[1:] + y[:-1])
area = np.sum(width*height)
```
